```python
### Part 2 - Data - Preprocessing
```

```python
#Guardar el % de nulos de cada columna en un diccionario 
dataPorcentNanDict = dict(df.isna().sum()/len(df)*100)
print(dataPorcentNanDict)
```

```python
#Retornar las columnas que contienen más de 10% de datos nulos
#Se utilizó diccionario en este caso
listDataNan = dict((k, v) for k, v in dataPorcentNanDict.items() if v >= 10)
print(listDataNan)
listKeyNan = list(listDataNan.keys())
print(listKeyNan)
```

```python
#Eliminar columnas que contienen más de 10% de datos nulos
df = df.drop(labels=listKeyNan, axis=1)
```

```python
print(df)
```

```python
#Eliminar las filas que contienen más de 5 valores nulos
df = df.dropna(thresh=144)
```

```python
print(df)
```

```python
#Guardar el tipo de cada columna en un diccionario 
TypeDict = dict(df.dtypes)
print(TypeDict)
```

```python
#Retornar los nombres de las columnas de tipo float64
#Se utilizó diccionario en este caso
listFloat64 = dict((k, v) for k, v in TypeDict.items() if v == "float64")
print(listFloat64)
listFloat = list(listFloat64.keys())
print(listFloat)
```

```python
df[listFloat] = df[listFloat].fillna(value=0)
```

```python
print(df)
```

```python
#Retorna el número de valores nulos de cada columna
df.isnull( ).sum( )
#En los siguientes resultados se puede ver que las columnas de tipo float64 ya no tienen nulos
```
